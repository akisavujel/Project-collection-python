```python
#IMPORT THE LIBRARY REQUIRED TO GENERATE CODE
import qrcode
```

```python
qr = qrcode.QRCode(
    version = 7, #Control the sieze of qr code
    box_size = 5, #Size of pixel in qr code
    border = 5 #Width of the border
)
```

```python
#Add the URL to generate QR code
data = input("Enter your URL: ") 
```

```python
#Embed link to QR code
qr.add_data(data)
```

```python
#Adjust size
qr.make(fit = True)
```

```python
#Set your desired colur
img = qr.make_image(fill = "pink",
              back_color = "pink")
```

```python
#Display the QR code
img
```
