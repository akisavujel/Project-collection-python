```python
import cv2 # Import OpenCV library
```

```python

# Load the default Haar cascade for frontal face detection
# cv2.data.haarcascades gives the full path to OpenCV's built-in cascades
face_cap = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# Open the webcam
# 0 -> default built-in camera (MacBook)
# 1 -> external camera if connected
video_cap = cv2.VideoCapture(1)
while True:

    # Read a frame from the camera
    ret, video_data = video_cap.read()
    # Convert frame to grayscale for faster face detection
    col = cv2.cvtColor(video_data, cv2.COLOR_BGR2GRAY)

    # Detect faces in the grayscale frame
    faces = face_cap.detectMultiScale(
        col,
        scaleFactor =1.1,
        minNeighbors = 5,
        minSize = (30,30),
        flags = cv2.CASCADE_SCALE_IMAGE)
    
    # Draw rectangles around detected faces
    for(x,y,w,h) in faces:
            cv2.rectangle(video_data,(x,y),(x+w,x+h),(0,255,0),2)

    # Show the live video with detected faces
    cv2.imshow("face_detection", video_data)

    # Press 'a' key to exit the loop immediately
    if cv2.waitKey(10) == ord("a"):
        break

video_cap.release()
```
